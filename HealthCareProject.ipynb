```python
import pandas as pd

data_health = pd.read_csv('/content/drowsiness_dataset.csv')
data_health.info()
```

```python
data_health.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
drow = data_health['Drowsiness']
print(drow.value_counts())
#plt.figure(figsize= (15,10))
sns.countplot(x = 'Drowsiness', data = data_health)
plt.show()
```

```python
data_health.columns
```

```python
# Scatter plot between heart Rate and drowsiness

sns.scatterplot(x='Heart_Rate', y ='Drowsiness',data = data_health)
plt.title('Scatter Plot of Heart Rate Vs Drowsiness')
plt.show()
```

```python
# Scatter plot between ppgGreen and drowsiness

sns.scatterplot(x='ppgGreen', y ='Drowsiness',data = data_health)
plt.title('Scatter Plot of ppgGreen Vs Drowsiness')
plt.show()
```

```python
#Plotting the correlation matrix
matrix = data_health.corr()
print(matrix)

#Plotting the correlation matrix
sns.heatmap(matrix, cmap = 'Greens', annot = True)
```

From the Heat map above , we can see that there is a very strong positive correlation between ppgRed and ppgIR.They are very much dependent on each other . as the ppgRed value increases the ppgIR value also increases and vice versa.

Drowsiness is having a negative correlation with Heart Rate,ppgRed and ppgIR and no correlation with ppgGreen.

  As the Heart Rate increases the Drowsines value decreases which means  when the heart Rate is high, the individual is very alert.

ppgGreen is having a positive  correlation with Heart Rate, ppgRed and ppgIR but no correlation with Drowsiness.

ppgIR is having a very strong correlation with ppgRed, positive correlation with ppgGreen nad Heart Rate but negative correlation with Drowsiness.

For an alert individual
the Heart Rate, ppgRed and the ppgIR value will be high.Not much dependent on ppgGreen value.


Heart Rate is having a positive correlation with the ppgRed, ppgGreen and ppgRed variables but a negative correlation with Drowsiness.

```python
sns.displot(data_health, x= 'ppgRed', kind = 'kde',fill =True)
plt.title('Distribuition plot with KDE for ppgRed')
plt.show()
```

```python
sns.displot(data_health, x="ppgIR", hue="Drowsiness", kind="kde", fill=True)
plt.title('Distribuition plot for ppgIR')
plt.show()
```

For alert individuals the ppgIR values are normally distribuited with its values between 5.5M and 6.25M

For drowsy individuals their ppgIR values vary between 6M and 6.5M

For significantly drowsy individuals the ppgIR values are between 4.75M and 5.25M
